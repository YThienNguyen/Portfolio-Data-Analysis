# Data Visualization using Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
tips_df=sns.load_dataset('tips');
```

```python
tips_df.head()
```

```python
sns.set_theme(); #apply the default theme
```

```python
sns.histplot(data=tips_df['total_bill']);
```

```python
tips_df["total_bill"].value_counts().sort_values(ascending=False);  ##check values the above plot
```

```python
sns.kdeplot(data=tips_df['total_bill']);  #sketch using kde style
```

```python
# specify total_bill based on time column

sns.displot(data=tips_df, x='total_bill',col='time', kde=True);
```

```python
tips_df.head()
```

```python
# showing which gender is giving more tips
sns.barplot(data=tips_df, x='sex', y='tip', estimator=np.mean);  
```

```python
tips_df['sex'].value_counts()
```

```python
#count how many customers are male and female
sns.countplot(data=tips_df, x='sex')
```

```python
# showing total_bill in each day, and how male & female perform theirs bill payments
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='sex', palette='Blues');
plt.legend(loc=0);
```

```python
tips_df.head()
```

```python
##showing correlation between between total_bill and tip based on smoker/non-smoker and time
tips_fg=sns.FacetGrid(data=tips_df, row='smoker', col='time');   #Create a class instance of Faacet Grid class
tips_fg.map(sns.scatterplot, 'total_bill', 'tip');
```

```python
#showing correlation between tip based and total_bill based on gender
#showing smoker 
kws=dict(s=100, edgecolor='b', alpha=.7)
new_fg=sns.FacetGrid(data=tips_df, col='sex',
                    col_order=['Female', 'Male'], hue='smoker',
                    palette='Set2',
                    height=4,
                    aspect=1.4)
new_fg.map(sns.scatterplot, 'total_bill', 'tip', **kws)
new_fg.add_legend();
```
